```python
import glob
import csv
import pandas as pd
```

```python
for file in glob.glob("read_datas_txt/*.txt"):
    print(file)
    li = []
    df = pd.read_csv(file,sep='|', encoding = "ISO-8859-1",skiprows=3,skipfooter =1 , engine = 'python',header=None,index_col=None)
    li.append(df)
    
```

```python
li
```

```python
frame = pd.concat(li, axis=0, ignore_index=True)
frame.to_excel("output.xlsx")
```
